#### Write a function density() that takes in a random variable and calculates the density function using `np.hist` and interpolation. The function should return two lists carrying x and y coordinates for plotting the density function

```python
def density(x):
    n, bins = np.histogram(x, 10, density = True)
    bins = [(bins[i] + bins[i+1])/2 for i in range(0, len(bins)-1)]
    return n, bins
    pass



# Generate test data and test the function - uncomment to run the test
# np.random.seed(5)
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 100)
n,bins = density(s)
plt.plot(bins,n, label = 'test')
plt.legend()
```

#### Add Overlapping density plots for male and female heights to the histograms plotted earlier

```python
# You code here 
fig = plt.figure()
ax_1 = fig.add_axes([0,0,1,1])
male_n, male_bins = density(Male_height)
female_n, female_bins = density(Female_height)
ax_1.hist(Male_height, density = True, bins = 10, alpha = 0.3)
ax_1.hist(Female_height, density = True, bins = 10, alpha = 0.3)
ax_1.plot(male_bins, male_n, color = 'green')
ax_1.plot(female_bins, female_n, color = 'purple')
```

#### Repeat above exerice for male and female weights

```python
# Your code here 
fig = plt.figure()
ax_1 = fig.add_axes([0,0,1,1])
male_n, male_bins = density(Male_weight)
female_n, female_bins = density(Female_weight)
ax_1.hist(Male_weight, density = True, bins = 10, alpha = 0.3)
ax_1.hist(Female_weight, density = True, bins = 10, alpha = 0.3)
ax_1.plot(male_bins, male_n, color = 'green')
ax_1.plot(female_bins, female_n, color = 'purple')
```
